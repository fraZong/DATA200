```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
```

1. Import CSV

```python
df = pd.read_csv('DATA200.csv')
```

2. Analyze Correlation

```python
corr_coeff = df['What is your age?'].corr(df['Do you believe that individuals should have the legal right to choose euthanasia or assisted suicide in cases of terminal illness or unbearable suffering?'])
print('Correlation Coefficient between Age & Question 1:', corr_coeff)
```

```python
corr_coeff = df['What is your age?'].corr(df['Do you think legalizing euthanasia or assisted suicide would alleviate or exacerbate concerns about end-of-life care?'])
print('Correlation Coefficient between Age & Question 2:', corr_coeff)
```

```python
corr_coeff = df['What is your age?'].corr(df['Should healthcare providers have the option to refuse participation in euthanasia or assisted suicide on moral or ethical grounds?'])
print('Correlation Coefficient between Age & Question 3:', corr_coeff)
```

```python
corr_coeff = df['What is your age?'].corr(df['Would you support the implementation of safeguards, such as mandatory waiting periods or psychological evaluations, for individuals seeking euthanasia or assisted suicide?'])
print('Correlation Coefficient between Age & Question 4:', corr_coeff)
```

```python
corr_coeff = df['What is your age?'].corr(df['Do you believe that legalizing euthanasia or assisted suicide could potentially lead to abuses or coercion?'])
print('Correlation Coefficient between Age & Question 5:', corr_coeff)
```

```python
corr_coeff = df['What is your age?'].corr(df['Should minors or individuals with mental illnesses be eligible for euthanasia or assisted suicide under certain circumstances?'])
print('Correlation Coefficient between Age & Question 6:', corr_coeff)
```

```python
corr_coeff = df['What is your age?'].corr(df['Should there be a distinction between voluntary euthanasia (where the patient consents) and non-voluntary euthanasia (where the patient is unable to provide consent)?'])
print('Correlation Coefficient between Age & Question 7:', corr_coeff)
```

# 3. Visualization

```python
sns.regplot(x='What is your age?', y='Would you support the implementation of safeguards, such as mandatory waiting periods or psychological evaluations, for individuals seeking euthanasia or assisted suicide?', data=df)
plt.title('Correlation between Age & Safeguard Implementation')
plt.xlabel('Age')
plt.ylabel('Safeguard Implementation Advocacy')
plt.show()
```

```python
Q1 = df['What is your age?'].quantile(0.25)
Q3 = df['What is your age?'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = df[(df['What is your age?'] < lower_bound) | (df['What is your age?'] > upper_bound)]

b_df = df[df['What is your age?'] <= upper_bound]
corr_coeff = b_df['What is your age?'].corr(b_df['Would you support the implementation of safeguards, such as mandatory waiting periods or psychological evaluations, for individuals seeking euthanasia or assisted suicide?'])
print('Correlation Coefficient between Age & Question 4:', corr_coeff)
```

```python
sns.regplot(x='What is your age?', y='Would you support the implementation of safeguards, such as mandatory waiting periods or psychological evaluations, for individuals seeking euthanasia or assisted suicide?', data=b_df)
plt.title('Correlation between Age & Safeguard Implementation (w/o outliers)')
plt.xlabel('Age')
plt.ylabel('Safeguard Implementation Advocacy')
plt.show()
```

```python
value_counts = df['What is your religious identity?'].fillna('n/a').value_counts()

plt.figure(figsize=(8, 6))
plt.pie(value_counts, labels=value_counts.index, autopct='%1.1f%%', startangle=70)
plt.title("Proportional Breakdown of the Participants\' Religious Identities")
plt.axis('equal')
plt.show()
```

```python
filtered_df = df
filtered_df['What is your religious identity?'] = filtered_df['What is your religious identity?'].replace({'Atheist': 1, 'Agnostic ': 1})
filtered_df['What is your religious identity?'] = filtered_df['What is your religious identity?'].replace(['Christian', 'Jewish', 'Muslim', 'Hindu', 'n/a', 'Buddhism', 'Bahai ', 'a mix of western and eastern philosophies (meditation, prayer)', 'Catholic ', 'Buddhist '], 0)

correlation_coefficient = filtered_df['What is your religious identity?'].corr(filtered_df['Do you believe that individuals should have the legal right to choose euthanasia or assisted suicide in cases of terminal illness or unbearable suffering?'])
print("Correlation Coefficient between Non-Religious People & Question 1:", correlation_coefficient)
correlation_coefficient = filtered_df['What is your religious identity?'].corr(filtered_df['Do you think legalizing euthanasia or assisted suicide would alleviate or exacerbate concerns about end-of-life care?'])
print("Correlation Coefficient between Non-Religious People & Question 2:", correlation_coefficient)
correlation_coefficient = filtered_df['What is your religious identity?'].corr(filtered_df['Should healthcare providers have the option to refuse participation in euthanasia or assisted suicide on moral or ethical grounds?'])
print("Correlation Coefficient between Non-Religious People & Question 3:", correlation_coefficient)
correlation_coefficient = filtered_df['What is your religious identity?'].corr(filtered_df['Would you support the implementation of safeguards, such as mandatory waiting periods or psychological evaluations, for individuals seeking euthanasia or assisted suicide?'])
print("Correlation Coefficient between Non-Religious People & Question 4:", correlation_coefficient)
correlation_coefficient = filtered_df['What is your religious identity?'].corr(filtered_df['Do you believe that legalizing euthanasia or assisted suicide could potentially lead to abuses or coercion?'])
print("Correlation Coefficient between Non-Religious People & Question 5:", correlation_coefficient)
correlation_coefficient = filtered_df['What is your religious identity?'].corr(filtered_df['Should minors or individuals with mental illnesses be eligible for euthanasia or assisted suicide under certain circumstances?'])
print("Correlation Coefficient between Non-Religious People & Question 6:", correlation_coefficient)
correlation_coefficient = filtered_df['What is your religious identity?'].corr(filtered_df['Should there be a distinction between voluntary euthanasia (where the patient consents) and non-voluntary euthanasia (where the patient is unable to provide consent)?'])
print("Correlation Coefficient between Non-Religious People & Question 7:", correlation_coefficient)
```

```python
sns.regplot(x='What is your religious identity?', y='Do you believe that individuals should have the legal right to choose euthanasia or assisted suicide in cases of terminal illness or unbearable suffering?', data=filtered_df)
plt.title('Correlation between Non-Religious Identity & Euthanasia Legality')
plt.xlabel('Non-Religious Identity')
plt.ylabel('Euthanasia Legality')
plt.show()
```

```python
filtered_df['What is your religious identity?'] = filtered_df['What is your religious identity?'].replace({1: 'Non-Religious'})
filtered_df['What is your religious identity?'] = filtered_df['What is your religious identity?'].replace(0, 'Religious')

# filtered_df['What is your religious identity?']
binary_variable_counts = filtered_df.groupby('What is your religious identity?')['Do you believe that individuals should have the legal right to choose euthanasia or assisted suicide in cases of terminal illness or unbearable suffering?'].value_counts().unstack()
ax = binary_variable_counts.plot(kind='bar', stacked=True)

plt.title('Clustered Bar Chart of Binary Correlation between Religious ID & Euthanasia Legality (Quantitative)')
plt.xlabel('Religious Identity')
plt.ylabel('Euthanasia Legality')
plt.xticks(rotation=0)
handles, labels = ax.get_legend_handles_labels()
handles = handles[::-1]
labels = labels[::-1]
plt.legend(handles, labels, title='Ordinal Legend\n(6 = Strong Yes, 1 = Strong No)', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
